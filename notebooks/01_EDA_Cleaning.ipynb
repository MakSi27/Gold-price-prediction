```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/MakSi27/Gold-price-prediction/main/gold_prices_10y.csv"
data = pd.read_csv(url)
```

```python
print(data)
```

```python
print(data.info())
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values('Date')
```

```python
print(data.dtypes)
```

```python
print(data.isna().sum())
```

```python
print("Duplicate rows ",data.duplicated().sum)
```

```python
print(data.describe())
```

```python
print("Unique values in each column:")
for col in data.columns:
    print(col, ":", data[col].nunique())
```

```python
for col in ['Year','Month', 'Day_of_Week','Quarter']:
  print("values count for ")
  print(data[col].value_counts())
  print("------------------------------------------")
```

```python
missing = data.isnull().sum()

missing.plot(kind='bar', figsize=(10,5))
plt.title("Missing Values per Column")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title("Heatmap")
corr_matrix = data.corr()
sn.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
