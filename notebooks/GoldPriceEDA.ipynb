# EDA

```python
sns.histplot(df['Close'],bins=50, kde=True)
plt.title('Distribution of Gold Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(df['Close'])
plt.title("Gold Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA_20'], label='MA 20')
plt.plot(df['MA_50'], label='MA 50')
plt.plot(df['MA_200'], label='MA 200')
plt.legend()
plt.title("Moving Averages")
plt.show()
```

```python
plt.plot(df['Volatility_20'])
plt.title("Gold Price Volatility (20 Day)")
plt.show()
```

### Seasonal (Monthly) Gold Price Trend

```python
sns.boxplot(x=df['Month'], y=df['Close'])
plt.title("Monthly Gold Price Distribution")
plt.show()
```

## Quaterly Gold Price trend

```python
sns.boxplot(x=df['Quarter'], y='Close', data=df)
plt.title("Quarterly Distribution of Gold Prices")
plt.xlabel("Quarter")
plt.ylabel("Gold Price")

plt.show()
```

## Day of week Gold price trend

```python
order = [0, 1, 2, 3, 4]
sns.boxplot(x='Day_of_Week', y='Close', data=df, order=order)

plt.title("Distribution of Gold Prices by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Gold Price")
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='plasma')
plt.title("Feature Correlation")
plt.show()
```

```python
sns.pairplot(df[['Open','High','Low','Close','Volume']])
plt.show()
```
