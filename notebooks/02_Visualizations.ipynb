```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/gold_prices_10y.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Close'], color='gold', linewidth=2)

plt.title("10-Year Gold Price Trend")
plt.xlabel("Year")
plt.ylabel("Gold Price")

plt.grid(True)

plt.show()
```

```python
plt.savefig("../reports/figures/gold_price_trend_10y.png")
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Close'], label="Price", color='gold')
plt.plot(df['Date'], df['MA_20'], label="MA 20", color='blue')
plt.plot(df['Date'], df['MA_50'], label="MA 50", color='green')
plt.plot(df['Date'], df['MA_200'], label="MA 200", color='red')

plt.title("Gold Price with Moving Averages")
plt.xlabel("Year")
plt.ylabel("Price")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Volatility_20'], color='purple')

plt.title("Gold Price Volatility (20 Day)")
plt.xlabel("Year")
plt.ylabel("Volatility")

plt.grid(True)

plt.show()
```
